```python
#Import dependencies 
import pandas as pd
import time
```

```python
#Read in CSV file

premise_start_time = time.time()

premise_errors=0

premise_count=0

csvFile="subset_a.csv"

csvDf=pd.read_csv(csvFile)

csvDf.head()
```

```python
#Create new dataframe with only wanted columns

newDf=csvDf[['id', 'task_id','adv_id','creat_type_cd','tags','age','city','city_rank','device_name','device_size','career','gender','residence','app_score','device_price','membership_life_duration','consume_purchase']]
newDf
```

```python
#Rename columns to be more descriptive

try:
    newColumns=['encounter_id', 'advertising_id','creative_type_id','membership_type']

    cleanedDf=newDf.rename(columns={'id':'encounter_id', 'adv_id':'advertising_id', 'creat_type_cd':'creative_type_id', 'membership_life_duration':'membership_type'})

except:
    
    premise_errors+=1
```

```python
#Store number of rows in variable

premise_count=cleanedDf.shape[0]
```

```python
#Create career table

careerIds=cleanedDf['career'].drop_duplicates().to_list()
careerList=['entertainment', 'arts', 'business', 'communications', 'farming', 'government', 'medicine', 'government','other']
careerTable=pd.DataFrame({'career_id':careerIds, 'careers':careerList}).set_index('career_id')
careerTable
```

```python
cityRankIds=cleanedDf['city_rank'].drop_duplicates().to_list()
cityRankList=['town', 'city', 'metropolis', 'megalopolis']
careerTable=pd.DataFrame({'city_rank_id':cityRankIds, 'city_rank':cityRankList}).set_index('city_rank_id')
careerTable
```

```python
cleanedDf.to_csv("cleanedData.csv")
```
